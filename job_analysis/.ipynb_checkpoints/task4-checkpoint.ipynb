### Справка по работе с API сервиса hh.ru.   

<b><font color="red">Внимание!</font> При работе с API не забывайте делать паузы между запросами, чтобы не задудосить сервер. Если вас заблокируют, это не будет являться уважительной причиной переноса дедлайна.</b>

Мы будем работать только с вакансиями. Для этого не требуется регистрироваться и получать токен. Ниже приведен краткий пример работы с API. Подробное описание работы с вакансиями, включая параметры запросов и формат ответа можно почитать в <a target="_blank" href="https://github.com/hhru/api/blob/master/docs/vacancies.md#search">документации</a>.

Например, мы хотим найти вакансии по запросу `Data Scientist` в Москве. Тогда первую страницу поиска из 10 вакансий на страницу мы можем получить с запроса к API:

```python
URL = 'https://api.hh.ru/vacancies'

params = {
    'text': "Data Scientist",
    'area': 1,
    'page': 0,
    'per_page': 10
}

req = requests.get(URL, params)
data = json.loads(req.content.decode())
```

Если все прошло успешно, полученный словарь будет иметь следующие ключи

```python
data.keys()
```

Сколько всего найдено вакансий

```python
data['found']
```

Количество страниц в результатах поиска

```python
data['pages']
```

Из результатов можем сделать удобную таблицу, причем в дальнейшем можно оставить только те колонки, которые необходимы для анализа.

```python
df = pd.json_normalize(data['items'])
df.head()
```

Для получения полного описания вакансии потребуется задать отдельный запрос, используя ее `id`.

```python
vacancy = df['id'].iloc[0]
vacancy_url = f'https://api.hh.ru/vacancies/{vacancy}'

req = requests.get(vacancy_url)
vacancy_info = json.loads(req.content.decode())
```

### Задача 1.   

Исследуем профессию Data Scientist. Найдите как можно больше вакансий по этой профессии в Москве. Учтите, что имеет смысл искать также по другим ключевым словам, например, `аналитик данных`.

```python
data = pd.DataFrame()
max_data = 2000
num_pages = 260
per_page = 100
vacancy_url = "https://api.hh.ru/vacancies"

for page in range(num_pages):
    if per_page * page >= max_data:
        per_page = max_data // page
    params = {
    'area' : 1,
    'page' : page,
    'per_page' : per_page,
    'text' : '"Data Scientist" OR "data scientist" OR "Aналитик Данных" OR "аналитик данных" OR "Data Analyst" OR "data analyst"'
    }
    try:
        request = requests.get(vacancy_url, params)
        data_from_page = json.loads(request.content.decode())
    
        df_from_page = pd.json_normalize(data_from_page['items'])
        data = pd.concat([data, df_from_page], ignore_index=True)
        sleep(0.4)
    except:
        pass
data.head()
```

```python
data.to_csv('vacancies')
data = pd.read_csv('vacancies')
```

```python
df1 = data[data['name'].apply(lambda a : 'data scientist' in a.lower())]
df2 = data[data['name'].apply(lambda a : 'data analyst' in a.lower())]
df3 = data[data['name'].apply(lambda a : 'аналитик данных' in a.lower())]
df = pd.concat([df1, df2, df3], ignore_index=True)
```

В полученную выборку некоторые вакансии могли попасть несколько раз. Удалите дубликаты.

```python
df = df.drop_duplicates('id')
df = df.dropna(how='all', axis=1)
df
```

Загрузите подробное описание каждой вакансии и создайте удобную для дальнейших действий таблицу данных.

```python
detailed_data = pd.DataFrame()
for item in df['id']:
    url = f'https://api.hh.ru/vacancies/{item}'
    request = requests.get(url)
    describtion = json.loads(request.content.decode())
    describtion_from_page = pd.json_normalize(describtion)
    detailed_data = pd.concat([detailed_data, describtion_from_page], ignore_index=True)
    sleep(0.4)
detailed_data
```

```python
detailed_data.to_csv('detailed_vacancie')
detailed_data = pd.read_csv('detailed_vacancie')
```

Теперь уберём лишние столбцы и приведем табличку в порядок.

```python
detailed_df = detailed_data.dropna(axis=1, how='all')
detailed_df = detailed_df.drop('branded_description', axis=1)
detailed_df = detailed_df.drop('description', axis=1)
for i in detailed_df.columns:
    if 'picture' in i:
        detailed_df = detailed_df.drop(i, axis=1)
    if 'logo' in i:
        detailed_df = detailed_df.drop(i, axis=1)
        
detailed_df
```

Теперь соединим две таблички

```python
for col in detailed_df.columns:
    if 'Unnamed' in col:
        detailed_df = detailed_df.drop(col, axis=1)

for col in df.columns:
    if 'Unnamed' in col:
        df = df.drop(col, axis=1)

result_data = pd.concat([df, detailed_df], axis=1)
result_data
```

Удалим почти полностью пустые столбцы, где пропусков 90%

```python
result_data = result_data.loc[:,~result_data.columns.duplicated()]
result_data = result_data.loc[:, result_data.isnull().mean() < .90]
result_data
```

**Вопрос 2.** Какие навыки чаще всего встречаются в вакансиях по данной специальности?

Для этого найдите соответствующее поле в данных из полного описания вакансий, проанализируйте его и составьте список навыков и количество упоминаний каждого. Визуализируйте полученную информацию по топ-15 навыков.

```python
all_skills = list()

for skills in result_data['key_skills']:
    if skills is np.nan:
        continue
    
    lst_skills = ast.literal_eval(skills)
    for item in lst_skills:
        all_skills.append(item['name'])
        
data_skills = pd.DataFrame()        
data_skills['key_skills'] = all_skills
data_skills
```

```python
plt.figure(figsize=(10, 10), dpi=80)
data_skills['key_skills'].value_counts()[:15].sort_values().plot.barh()
plt.title('Самые популярные навыки');
plt.xlabel('Популярность');
plt.ylabel('Навык');
```

**Вопрос 3.** Какую зарплату готовы платить работодатели? Соберите некоторым образом статистику и постройте гистограмму.

При работе с данными о заработной плате обратите внимание на валюту и <a target="_blank" href="https://yandex.ru/search/?text=gross%2Fnet&clid=1955455&win=446&lr=213&noreask=1">gross/net</a>. Постоянно подгружать курс валюты не требуется, достаточно фиксировать какой-то один более менее актуальный.

Посмотрим, какие признаки есть в табличке

```python
result_data.columns
```

```python
salary = result_data[['salary.from', 'salary.to', 'salary.currency']]
salary = salary[(salary['salary.from'].notna() | salary['salary.to'].notna()) & salary['salary.currency'].notna()]
salary
```

Уберём пропуски, заменив их на минимум и максимум в столбце.

```python
min = salary['salary.from'].min()
max = salary['salary.to'].max()
salary['salary.to'] = salary['salary.to'].replace(np.nan, max)
salary['salary.from'] = salary['salary.from'].replace(np.nan, min)
salary
```

Загрузим табличку с курсом валют.

```python
url = 'https://api.hh.ru/dictionaries'
request = requests.get(url)
corrency = json.loads(request.content.decode())
corrency_data = pd.DataFrame(columns = ['code', 'rate'])
i = 0
for rate in corrency['currency']:
    corrency_data.loc[i] = [rate['code'], rate['rate']]
    i += 1
corrency_data
```

Добавим столбец курса валют к табличке salary.

```python
def toRate(elem):
    elem['rate'] = float(corrency_data[corrency_data['code'] == elem['salary.currency']]['rate'])
    return elem
```

```python
salary = salary.apply(lambda x: toRate(x), axis=1)
salary
```

```python
salary['salary.from'] /= salary['rate']
salary['salary.to'] /= salary['rate']
salary = salary.drop('rate', axis=1)
salary = salary.drop('salary.currency', axis=1)
salary['salary.from'] = salary['salary.from'].apply(lambda x : int(x))
salary['salary.to'] = salary['salary.to'].apply(lambda x : int(x))
salary
```

```python
salary.describe()
```
